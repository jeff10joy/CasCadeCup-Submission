```python
import pandas as pd
```

```python
df=pd.read_csv("train_age_dataset.csv")
```

```python
df.shape
```

```python
df.head(100)
```

```python
df.drop(['Unnamed: 0', 'userId'], axis = 1, inplace = True)
```

```python
c = df.astype(bool).sum(axis=0)
c
```

```python
to_drop = []
for feature in c.index:
    if c[feature]<100000:
        to_drop+=[feature]
to_drop
```

```python
def impute(feature):
    global df
    data = df[[feature, 'age_group']].copy().reset_index()
    data = data[data[feature]!=0]
    class_avgs = []
    for i in range(4):
        values = list(data[data['age_group'] == 1][feature].values)
        class_avgs += [sum(values)/len(values)]
    indices = df[df[feature] == 0].index
    for index in indices:
        df.at[index, feature] = class_avgs[df.at[index, 'age_group'] - 1]

total_features = list(set(df.columns) - set(to_drop))
for feature in total_features:
    if c[feature]<len(df):
        impute(feature)
```

```python
df.to_csv("clean_train.csv", index = False)
```

```python
from catboost import CatBoostClassifier
import xgboost as xgb
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df[df.columns[:-1]],df[df.columns[-1]], test_size = 0.2, random_state=2020)
x_train,x_test,y_train,y_test = x_train.reset_index(drop = True),x_test.reset_index(drop = True),y_train.reset_index(drop = True),y_test.reset_index(drop = True)
print(x_train.shape)
print(y_train.shape)
x_train.tail()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
corr=df[df.columns[2:-1]].corr()
plt.figure(figsize=(20,20))
sns.heatmap(corr, annot = True)
```

```python

features=['slot2_trails_watched_per_day','weekdays_trails_watched_per_day','weekends_trails_watched_per_day','slot4_trails_watched_per_day','slot3_trails_watched_per_day','content_views']
x = x_train[features]
pca=PCA(n_components=2)
pca_obj=pca.fit(x)
x=pca_obj.transform(x)
len(x)
```

```python

df2=pd.DataFrame(x)
print(df2.shape)
x_train.drop(features + to_drop, axis=1,inplace=True)
print(x_train.shape)

data=pd.concat([x_train,df2],axis=1)

print(data.shape)
data.tail()
```

```python
#model = RandomForestClassifier(n_jobs = -1, max_features = "log2", n_estimators = 100)
#model = CatBoostClassifier(iterations=300,random_seed=63,learning_rate=0.7,custom_loss=['AUC', 'Accuracy'])
model=xgb.XGBClassifier()
model.fit(data, y_train)
```

```python
y=x_test[features]
y=pca_obj.transform(y)
df3=pd.DataFrame(y)
x_test.drop(features + to_drop, axis=1, inplace = True)
x_test=pd.concat([x_test,df3],axis=1)
```

```python
from sklearn.metrics import f1_score
y_pred = model.predict(x_test)
f1_score(y_test,y_pred, average = 'weighted')
```
