```python
from catboost import CatBoostClassifier
import xgboost as xgb
from sklearn.decomposition import PCA

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df[df.columns[2:-1]],df[df.columns[-1]], test_size = 0.2, random_state=2020)
print(x_train.shape)
print(y_train.shape)
```

```python
import seaborn as sns
corr=df[df.columns[2:-1]].corr()
corr
```

```python

features=['slot2_trails_watched_per_day','creations','num_of_comments','weekdays_trails_watched_per_day','weekends_trails_watched_per_day','slot4_trails_watched_per_day','slot3_trails_watched_per_day','content_views']
x = x_train.loc[:, features].values
pca=PCA(n_components=2)
pca_obj=pca.fit(x)
x=pca_obj.transform(x)
len(x)
```

```python

df2=pd.DataFrame(x)
print(len(df2))
x_train.drop(features,axis=1,inplace=True)
print(len(x_train))
df2=df2.reset_index(drop=False)

data=pd.concat([x_train,df2],axis=1)

print(len(data))
```

```python
x_train.tail()
```

```python

#model = CatBoostClassifier(iterations=300,random_seed=63,learning_rate=0.7,custom_loss=['AUC', 'Accuracy'])
model=xgb.XGBClassifier()
model.fit(
    data, y_train
)
```

```python
y=x_test.loc[:,features].values
y=pca.fit_transform(y)
df3=pd.DataFrame(y)
x_test.drop(features,axis=1)
x_test=pd.concat([x_test,df3],axis=1)
```

```python
from sklearn.metrics import f1_score
y_pred = model.predict(x_test)
f1_score(y_test,y_pred, average = 'weighted')
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
clf = AdaBoostClassifier(n_estimators=10000, random_state=0)
clf.fit(x_train, y_train)
```

```python
y_pred=clf.predict(x_test)
f1_score(y_test,y_pred,average='weighted')
```
